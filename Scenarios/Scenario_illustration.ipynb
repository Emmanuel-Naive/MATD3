```python
import numpy as np
import matplotlib.pyplot as plt
import math
colorset = ['b', 'r', 'g', 'y', 'c', 'm']
t = 20
```

# 1shipsM
## parameters

```python
ships_num = 1

ships_init = np.zeros((ships_num, 2))
ships_goal = np.zeros((ships_num, 2))
ships_speed = np.zeros((ships_num, 1))
ships_head = np.zeros((ships_num, 1))

ships_init[0, :] = np.array([5000, 0])
ships_goal[0, :] = np.array([0, 10000])
ships_speed[0] = 20
ships_head[0] = 90
```

```python
plt.figure(figsize=(5,5))
for i in range(ships_num):
    plt.scatter(ships_init[i,0], ships_init[i,1], color=colorset[i], label='ship{}_initial_position'.format(i+1))
    plt.scatter(ships_goal[i,0], ships_goal[i,1], marker="x", color=colorset[i], label='ship{}_goal'.format(i+1))
    xt = ships_init[i,0] + (ships_speed[i] * math.cos(math.radians(ships_head[i])) * t)
    yt = ships_init[i,1] + (ships_speed[i] * math.sin(math.radians(ships_head[i])) * t)
    plt.plot([ships_init[i,0], xt], [ships_init[i,1], yt], color=colorset[i], label='ship{}_heading_angle'.format(i+1))
#     plt.plot([ships_init[i,0], ships_goal[i,0]], [ships_init[i,1], ships_goal[i,1]], "--", alpha=0.2, color=colorset[i], label='ship{}_shortest_path_to_goal'.format(i+1))
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.title('Scenario: 1shipsM')
plt.grid()
```

```python
ships_num = 4

ships_init = np.zeros((ships_num, 2))
ships_goal = np.zeros((ships_num, 2))
ships_speed = np.zeros((ships_num, 1))
ships_head = np.zeros((ships_num, 1))

ships_init[0, :] = np.array([2000, 0])
ships_goal[0, :] = np.array([-2000, 0])
ships_speed[0] = 20
ships_head[0] = 180

ships_init[1, :] = np.array([-2000, 0])
ships_goal[1, :] = np.array([2000, 0])
ships_speed[1] = 20
ships_head[1] = 0

ships_init[2, :] = np.array([-1500, 0])
ships_goal[2, :] = np.array([1500, 0])
ships_speed[2] = 15
ships_head[2] = 0

ships_init[3, :] = np.array([-1000, 0])
ships_goal[3, :] = np.array([1000, 0])
ships_speed[3] = 10
ships_head[3] = 0
```

## plot

```python
plt.figure(figsize=(5,5))
for i in range(ships_num):
    plt.scatter(ships_init[i,0], ships_init[i,1], color=colorset[i], label='ship{}_initial_position'.format(i+1))
    plt.scatter(ships_goal[i,0], ships_goal[i,1], marker="x", color=colorset[i], label='ship{}_goal'.format(i+1))
    xt = ships_init[i,0] + (ships_speed[i] * math.cos(math.radians(ships_head[i])) * t)
    yt = ships_init[i,1] + (ships_speed[i] * math.sin(math.radians(ships_head[i])) * t)
    plt.plot([ships_init[i,0], xt], [ships_init[i,1], yt], color=colorset[i], label='ship{}_heading_angle'.format(i+1))
#     plt.plot([ships_init[i,0], ships_goal[i,0]], [ships_init[i,1], ships_goal[i,1]], "--", alpha=0.2, color=colorset[i], label='ship{}_shortest_path_to_goal'.format(i+1))
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.ylim(-500, 500)
plt.title('Scenario: 4ships_H4O3')
plt.grid()
```
